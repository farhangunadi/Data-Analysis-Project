### Data Analysis
## Adidas Sales Dataset.

An Adidas sales dataset is a collection of data that includes information on the sales of Adidas products. This type of dataset may include details such as the number of units sold, the total sales revenue, the location of the sales, the type of product sold, and any other relevant information.

*Original dataset can be downloaded from the **[Kaggle.com](https://www.kaggle.com/datasets/heemalichaudhari/adidas-sales-dataset)** website*

## Data Loaded

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data from directory
df_adidas = pd.read_excel('./adidas_sales.xlsx')
df_adidas.head(10)
```

### Data Cleaning and Processing

```python
#remove first three row and change the name column with the fourth row
df = df_adidas.drop(df_adidas.index[0:3])
df.columns = df.iloc[0]
df = df[1:]
df.head()
```

```python
#remove the index
df = df.reset_index(drop=True)
df
```

```python
#remove the first column
df = df.iloc[:, 1:]
df
```

```python
df.info()
```

```python
convert_to_int = ['Price per Unit','Units Sold','Total Sales']
convert_to_float = ['Operating Profit','Operating Margin']

for col in convert_to_int:
    df[col] = df[col].astype(int)
    
for col in convert_to_float:
    df[col] = df[col].astype(float)

df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])

df.info()
```

```python
df.info()
```

```python
df['Year'] = df['Invoice Date'].dt.year
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

```python
df.head()
```

### Data Visulization

```python
total_sales = df.groupby('Region')['Total Sales'].sum()
total_sales
```

```python
#Show Total Sales by Region
sns.catplot(x='Region',y='Total Sales',data=df,kind='bar',palette="ch:.25")
```

```python
#Show total Operating Profit by Region
sns.catplot(x='Year',y='Operating Profit',data=df,kind='bar',palette="ch:.25")
```

```python
#Show Total Units Sold by Gender Product
def product(type):
    if type in ["Men's Street Footwear","Men's Athletic Footwear","Men's Apparel"]:
        return "Men's Product"
    else:
        return "Women's Product"

#Grouped product by gender
product_grouped = df.groupby(df['Product'].apply(product))
unit_sold = product_grouped['Units Sold'].mean()
result_df = unit_sold.reset_index()

#Bar Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Product', y='Units Sold', data=result_df,palette="ch:.25")
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.show()
```

```python
#Show Total Product Sales Every month
sns.lineplot(x="Year", y="Total Sales",hue="Product", data=df, errorbar=('ci', False))
```

```python
#Show total transaction in every retail
sns.countplot(x='Retailer', data=df,palette="ch:.25" )
```
