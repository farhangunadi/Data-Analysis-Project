```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Financials.csv')
df.head(10)
```

Hapus White Space pada  nama kolom

```python
# Remove spaces from the first and last characters of column names
new_columns = [col.strip() for col in df.columns]
# Update the column names in the DataFrame
df.columns = new_columns
print(df.columns)
```

Hapus '$' dan ','  pada tiap data

```python
# Specify the character or substring you want to remove
character_to_remove = '$'
character_to_remove2 = ','


# Iterate over the columns and apply .str.replace() to each column
for column in df.columns:
  if df[column].dtype == 'object': # Only apply to object (string) columns
    df[column] = df[column].str.replace(character_to_remove,'')
    df[column] = df[column].str.replace(character_to_remove2,'')

df.head(10)
```

```python
df['Discounts'] = df['Discounts'].str.replace('-','0')
df['Profit'] = df['Discounts'].str.replace('-','0')
```

Ganti data type dari object ke int untuk setiap data angka

```python
df.dtypes
# Specify the column names you want to convert to int
columns_to_convert = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales ','Discounts','Sales','COGS','Profit']

# Convert the specified columns' data type to int
for column in columns_to_convert:
  df[column] = df[column].str.replace(' ', '').astype(float)

df.head(10)
```

Pembulatan angka di belakang koma di kolom Units Sold

```python
df['Units Sold'] = np.ceil(df['Units Sold'])

df.head(10)
```

```python
df['Units Sold'] = df['Units Sold'].astype(int)

df.head(10)
```

Save data baru kedalam csv file

```python
output_file = 'Financials_cleandata.csv'

df.to_csv(output_file, index=False)

print(f"DataFrame saved to {output_file}")
```

# **Data Visualization**

```python
df_fin = pd.read_csv('Financials_cleandata.csv')
df_fin.info()
```

```python
#change 'Date' datatype to date
df_fin['Date'] = pd.to_datetime(df_fin['Date'])
df_fin.info()
```

```python
df_fin['Year'] = pd.to_datetime(df_fin['Year'], format='%Y')
```

```python
grouped = df_fin.groupby(['Year', 'Country']).sum().reset_index()

pivot_data = grouped.pivot(index='Year', columns='Country', values='Sales')
pivot_data.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Sales by Year Categorized by Country')

# Display the plot
plt.tight_layout()
plt.show()
```
